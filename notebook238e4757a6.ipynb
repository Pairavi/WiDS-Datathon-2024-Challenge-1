```python
import pandas as pd
import numpy as np
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the dataset
train_df = pd.read_csv('/kaggle/input/widsdatathon2024-challenge1/training.csv')
test_df = pd.read_csv("/kaggle/input/widsdatathon2024-challenge1/test.csv")

train = TabularDataset(train_df)
test = TabularDataset(test_df)


# Initialize AutoGluon with GPU support
automl = TabularPredictor(label='DiagPeriodL90D', problem_type='regression', eval_metric='root_mean_squared_error')
automl.fit(train, presets='best_quality')
```

```python
automl.leaderboard()
```

```python
prediction = automl.predict(test)
```

```python
prediction
```

```python
submission_df = pd.DataFrame({'patient_id': test_df['patient_id'], 'DiagPeriodL90D': prediction})
```

```python
submission_df.head(50)
```

```python
import csv

# Specify the file path
file_path = 'submission.csv'

# Attempt to write the DataFrame to a CSV file
try:
    # Open the file for writing
    with open(file_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        
        # Write the header row
        writer.writerow(submission_df.columns)
        
        # Write each row of data
        for _, row in submission_df.iterrows():
            writer.writerow(row)
    
    print("CSV file saved successfully.")
except Exception as e:
    print("Error occurred while saving CSV file:", e)
```
